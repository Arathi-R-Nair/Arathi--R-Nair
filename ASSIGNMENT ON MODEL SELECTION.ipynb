```python
df.drop('Name',axis=1,inplace=True)
df.drop('Cabin',axis=1,inplace=True)
df.dropna(inplace=True)
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean)
```

```python
df['Age']=df['Age'].fillna(df['Embarked'].mode()[0])
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
df['Ticket']=le.fit_transform(df['Ticket'])
df['Embarked']=le.fit_transform(df['Embarked'])
df.head()
```

```python
x=df.drop('Survived',axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
k_metrics=[]
neighbor=np.arange(3,15)
for k in neighbor:
    knn=KNeighborsClassifier(n_neighbors=k,metric='euclidean')
    knn.fit(x_train,y_train)
    y_predict=knn.predict(x_test)
    acc=accuracy_score(y_test,y_predict)
    k_metrics.append(acc)
    
k_metrics
```

# SVM

```python
from sklearn.svm import SVC
```

```python
svm_cls=SVC(kernel='linear')
svm_cls=svm_cls.fit(x_train,y_train)
y_pred_svm=svm_cls.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred_svm)
```

```python
print('Accuracy=',accuracy_score(y_test,y_pred_svm))
print('Precision=',precision_score(y_test,y_pred_svm))
print('recall=',recall_score(y_test,y_pred_svm))
print('f1=',f1_score(y_test,y_pred_svm))
```

# K-FOLD

```python
from sklearn.model_selection import KFold
kfold_Validator=KFold(10)
for train_index,test_index in kfold_Validator.split(x,y):
    print('Training Index:',train_index)
    print('Testing Index:',test_index)
```

```python
from sklearn.model_selection import cross_val_score
cv_result=cross_val_score(svm_cls,x,y,cv=kfold_Validator)
```

```python
cv_result
```

```python
np.mean(cv_result)
```

# Stratified Cross Validation

```python
from sklearn.model_selection import StratifiedKFold
skfold_validator=StratifiedKFold(n_splits=10)
```

```python
for train_index,test_index in skfold_validator.split(x,y):
    print('Training Index:',train_index)
    print('Testing Index:',test_index)
```

```python
cv_result=cross_val_score(svm_cls,x,y,cv=skfold_validator)
```
